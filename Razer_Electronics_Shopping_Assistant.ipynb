```python
import tkinter as tk
from tkinter import messagebox
from experta import Fact, Rule, KnowledgeEngine, NOT, AND, MATCH

# Define a fact to represent user preferences
class UserPreferences(Fact):
    pass

# Define the expert system
class ShoppingAssistant(KnowledgeEngine):

    # Smartphone Recommendations
    @Rule(AND(UserPreferences(price='low'), UserPreferences(category='smartphone'), UserPreferences(feature='battery_life'), UserPreferences(brand='oppo')))
    def recommend_smartphone1(self):
        self.recommendation = "Oppo A54 - Budget Smartphone with long battery life."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='smartphone'), UserPreferences(feature='performance'), UserPreferences(brand='samsung')))
    def recommend_smartphone2(self):
        self.recommendation = "Samsung Galaxy A52 - Mid-Range Smartphone with great performance."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='smartphone'), UserPreferences(feature='camera'), UserPreferences(brand='apple')))
    def recommend_smartphone3(self):
        self.recommendation = "iPhone 13 - Premium Smartphone with excellent camera quality."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='smartphone'), UserPreferences(feature='waterproof'), UserPreferences(brand='google')))
    def recommend_smartphone4(self):
        self.recommendation = "Google Pixel 5 - Mid-Range Waterproof Smartphone."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='smartphone'), UserPreferences(feature='gaming'), UserPreferences(brand='asus')))
    def recommend_smartphone5(self):
        self.recommendation = "Asus ROG Phone 5 - High-End Gaming Smartphone."

    # Laptop Recommendations
    @Rule(AND(UserPreferences(price='low'), UserPreferences(category='laptop'), UserPreferences(feature='performance'), UserPreferences(brand='acer')))
    def recommend_laptop1(self):
        self.recommendation = "Acer Aspire 5 - Budget Laptop with decent performance."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='laptop'), UserPreferences(feature='gaming'), UserPreferences(brand='dell')))
    def recommend_laptop2(self):
        self.recommendation = "Dell G5 15 - Mid-Range Gaming Laptop."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='laptop'), UserPreferences(feature='ultra_light'), UserPreferences(brand='macbook')))
    def recommend_laptop3(self):
        self.recommendation = "MacBook Air - Premium Ultra-Light Laptop."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='laptop'), UserPreferences(feature='business'), UserPreferences(brand='hp')))
    def recommend_laptop4(self):
        self.recommendation = "HP ProBook 450 - Mid-Range Business Laptop."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='laptop'), UserPreferences(feature='2_in_1'), UserPreferences(brand='lenovo')))
    def recommend_laptop5(self):
        self.recommendation = "Lenovo Yoga 9i - Premium 2-in-1 Laptop."

    # Tablet Recommendations
    @Rule(AND(UserPreferences(price='low'), UserPreferences(category='tablet'), UserPreferences(feature='budget'), UserPreferences(brand='amazon')))
    def recommend_tablet1(self):
        self.recommendation = "Amazon Fire HD 10 - Budget Tablet."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='tablet'), UserPreferences(feature='performance'), UserPreferences(brand='samsung')))
    def recommend_tablet2(self):
        self.recommendation = "Samsung Galaxy Tab S6 - Mid-Range Tablet with great performance."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='tablet'), UserPreferences(feature='premium'), UserPreferences(brand='apple')))
    def recommend_tablet3(self):
        self.recommendation = "iPad Pro - Premium Tablet with excellent features."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='tablet'), UserPreferences(feature='stylus'), UserPreferences(brand='microsoft')))
    def recommend_tablet4(self):
        self.recommendation = "Microsoft Surface Pro 7 - Mid-Range Tablet with stylus support."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='tablet'), UserPreferences(feature='gaming'), UserPreferences(brand='razer')))
    def recommend_tablet5(self):
        self.recommendation = "Razer Edge - High-End Gaming Tablet."

    # Smartwatch Recommendations
    @Rule(AND(UserPreferences(price='low'), UserPreferences(category='smartwatch'), UserPreferences(feature='fitness'), UserPreferences(brand='xiaomi')))
    def recommend_smartwatch1(self):
        self.recommendation = "Xiaomi Mi Band 6 - Budget Fitness Smartwatch."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='smartwatch'), UserPreferences(feature='style'), UserPreferences(brand='fitbit')))
    def recommend_smartwatch2(self):
        self.recommendation = "Fitbit Versa 3 - Stylish Mid-Range Smartwatch."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='smartwatch'), UserPreferences(feature='premium'), UserPreferences(brand='apple')))
    def recommend_smartwatch3(self):
        self.recommendation = "Apple Watch Series 7 - Premium Smartwatch with advanced features."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='smartwatch'), UserPreferences(feature='health'), UserPreferences(brand='garmin')))
    def recommend_smartwatch4(self):
        self.recommendation = "Garmin Forerunner 245 - Mid-Range Health-Focused Smartwatch."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='smartwatch'), UserPreferences(feature='luxury'), UserPreferences(brand='tag_heuer')))
    def recommend_smartwatch5(self):
        self.recommendation = "TAG Heuer Connected - Luxury Smartwatch."

    # Headphone Recommendations
    @Rule(AND(UserPreferences(price='low'), UserPreferences(category='headphone'), UserPreferences(feature='budget'), UserPreferences(brand='jbl')))
    def recommend_headphone1(self):
        self.recommendation = "JBL Tune 500 - Budget Headphones."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='headphone'), UserPreferences(feature='noise_cancellation'), UserPreferences(brand='sony')))
    def recommend_headphone2(self):
        self.recommendation = "Sony WH-CH710N - Mid-Range Noise-Cancelling Headphones."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='headphone'), UserPreferences(feature='premium'), UserPreferences(brand='bose')))
    def recommend_headphone3(self):
        self.recommendation = "Bose 700 - Premium Noise-Cancelling Headphones."

    @Rule(AND(UserPreferences(price='medium'), UserPreferences(category='headphone'), UserPreferences(feature='gaming'), UserPreferences(brand='razer')))
    def recommend_headphone4(self):
        self.recommendation = "Razer Kraken X - Mid-Range Gaming Headphones."

    @Rule(AND(UserPreferences(price='high'), UserPreferences(category='headphone'), UserPreferences(feature='audiophile'), UserPreferences(brand='sennheiser')))
    def recommend_headphone5(self):
        self.recommendation = "Sennheiser HD 800S - Premium Audiophile Headphones."

    # No Recommendation Handling
    @Rule(NOT(AND(UserPreferences(price=MATCH.price), UserPreferences(category=MATCH.category), UserPreferences(feature=MATCH.feature), UserPreferences(brand=MATCH.brand))))
    def no_recommendation(self):
        self.recommendation = "We couldn't find an exact match for your preferences, but how about checking out our bestseller in your chosen category? Let us know if you're interested!"

# Function to run the recommendation engine
def run_recommendation():
    price = price_var.get()
    category = category_var.get()
    feature = feature_var.get()
    brand = brand_var.get()
    color = color_var.get()

    # Create an instance of the expert system
    engine = ShoppingAssistant()
    engine.reset()  # Reset the engine to clear any previous state

    # Declare the user's preferences as facts
    engine.declare(UserPreferences(price=price, category=category, feature=feature, brand=brand, color=color))

    # Run the expert system to generate recommendations
    engine.run()

    # Print the final recommendation
    if hasattr(engine, 'recommendation'):
        user_response = messagebox.askyesno("Recommendation", f"{engine.recommendation}\nWould you like to explore this option?")
        if user_response:
            messagebox.showinfo("Thank You", "Great choice! We hope you enjoy your shopping experience with Razer.")
        else:
            messagebox.showinfo("Suggestion", "Feel free to adjust your preferences and try again.")
    else:
        messagebox.showinfo("Recommendation", "No recommendations found based on your preferences.")

# Create the main window
root = tk.Tk()
root.title("Razer - Your Electronics Shopping Assistant")
root.geometry("500x500")
root.configure(bg="#f0f0f0")

# Welcome Message
messagebox.showinfo("Welcome", "Welcome to Razer - Your personalized electronics shopping assistant!")

# Create and place labels and entry fields
tk.Label(root, text="Select Price Range:", bg="#f0f0f0").pack(pady=5)
price_var = tk.StringVar(value="low")
tk.OptionMenu(root, price_var, "low", "medium", "high").pack(pady=5)

category_options = {
    "smartphone": ["oppo", "samsung", "apple", "google", "asus"],
    "laptop": ["acer", "dell", "macbook", "hp", "lenovo"],
    "tablet": ["amazon", "samsung", "apple", "microsoft", "razer"],
    "smartwatch": ["xiaomi", "fitbit", "apple", "garmin", "tag_heuer"],
    "headphone": ["jbl", "sony", "bose", "razer", "sennheiser"]
}

category_features = {
    "smartphone": ["battery_life", "performance", "camera", "waterproof", "gaming"],
    "laptop": ["performance", "gaming", "ultra_light", "business", "2_in_1"],
    "tablet": ["budget", "performance", "premium", "stylus", "gaming"],
    "smartwatch": ["fitness", "style", "premium", "health", "luxury"],
    "headphone": ["budget", "noise_cancellation", "premium", "gaming", "audiophile"]
}

# Update brand and feature dropdowns based on category
def update_options(*args):
    category = category_var.get()
    
    # Update feature menu
    feature_menu['menu'].delete(0, 'end')
    new_features = category_features.get(category, [])
    for feature in new_features:
        feature_menu['menu'].add_command(label=feature, command=tk._setit(feature_var, feature))
    feature_var.set(new_features[0] if new_features else "")

    # Update brand menu
    brand_menu['menu'].delete(0, 'end')
    new_brands = category_options.get(category, [])
    for brand in new_brands:
        brand_menu['menu'].add_command(label=brand, command=tk._setit(brand_var, brand))
    brand_var.set(new_brands[0] if new_brands else "")

# Category selection
category_var = tk.StringVar(value="smartphone")
category_var.trace("w", update_options)
tk.Label(root, text="Select Category:", bg="#f0f0f0").pack(pady=5)
tk.OptionMenu(root, category_var, *category_options.keys()).pack(pady=5)

# Feature selection
feature_var = tk.StringVar(value="battery_life")
tk.Label(root, text="Select Feature Preference:", bg="#f0f0f0").pack(pady=5)
feature_menu = tk.OptionMenu(root, feature_var, "")
feature_menu.pack(pady=5)

# Brand selection
brand_var = tk.StringVar(value="oppo")
tk.Label(root, text="Select Brand Preference:", bg="#f0f0f0").pack(pady=5)
brand_menu = tk.OptionMenu(root, brand_var, "")
brand_menu.pack(pady=5)

# Color selection
color_var = tk.StringVar(value="black")
tk.Label(root, text="Select Color Preference:", bg="#f0f0f0").pack(pady=5)
tk.OptionMenu(root, color_var, "black", "white", "blue", "red", "green", "silver").pack(pady=5)

# Button to run the recommendation engine
tk.Button(root, text="Get Recommendation", command=run_recommendation, bg="#0078D7", fg="white").pack(pady=20)

# Exit Message
def on_exit():
    messagebox.showinfo("Goodbye", "Thank you for choosing Razer. Have a great day!")
    root.destroy()

root.protocol("WM_DELETE_WINDOW", on_exit)

# Initialize the options based on the default category
update_options()

# Run the application
root.mainloop()
```
